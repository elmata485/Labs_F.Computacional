# Parte II

Primero importo las librerías que voy a usar

````python
import numpy as np
import matplotlib.pyplot as plt
````

Ahora defino la función que está igualada a la derivada de ```y```

````python
def f(x,y):

    return 2 * (1-y) - np.exp(-4*x)
````

Aquí defino la función Runge-Kutta de 4to orden, lo hice siguiendo la explicación del método que dió el profe en el laboratorio 3.

````python
def rk4(f,x_n,y_n,h):

    k_1 = f( x_n , y_n )

    k_2 = f( x_n+(h/2) , y_n+ ((h/2)*k_1) )

    k_3 = f( x_n+(h/2), y_n+ ((h/2)*k_2) )

    k_4 = f( x_n+h , y_n+(h*k_3) )

    return y_n + (h/6) * (k_1 + 2*k_2 + 2*k_3 + k_4) 
````

Ahora declaro valores iniciales e itero la función rk4.

````python
N = 1000  # Número de divisiones del intervalo

x_i = 0   # Valor inicial del intervalo de x
x_f = 2   # Valor final del intervalo de x

y_i = 1   # Condición inicial

h = (x_f - x_i)/N   # Este es el tamaño de las divisiones del intervalo

x = np.zeros(N+1)   # Aquí declaro un arreglo de ceros para recorrer el intervalo [0,2] en pasos de "h"
y = np.zeros(N+1)   # Aquí declaro un arreglo de ceros para guardar los resultados de las iteraciones de rk4

x[0] = x_i  # Inicializo la primer entrada de x con el valor inicial de x_i
y[0] = y_i  # Inicializo la primer entrada de y con el valor inicial de y_i

for i in range(0,N): 

    x[i+1] = x[i] + h  # Recorro el intervalo en pasos de h
    
    y[i+1] = rk4(f,x[i],y[i],h)   # Guardo el y_n+1 que me devuelve rk4
````

Ahora grafico ```y``` en función de ```x```

````python
plt.figure()

plt.plot(x, y)

plt.show()
````
